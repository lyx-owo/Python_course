假設班上10位同學的數學成績是90,80,88,84,85,87,88,89,100

假設班上10位同學的數學成績80,90,85,75,70,65,88,98,100
75分改成78
印出最高最低
在尾端加上50,99
由大到小
由小到大



```python
sc=[80,90,85,75,70,65,88,98,100]
sc[3]=sc[3]+3
print("75分改成78",sc)
print("目前分數=" ,sc)
sc.append(50)
print("目前分數=" ,sc)
sc.append(99)
print("最高分=", max(sc[0:9]))
print("最低分=", min(sc[0:9]))
print("總分=", sum(sc[0:9]))

print("目前串列內容",sc)
print("使用sort()由大到小")
sc.sort(reverse=True)
print("由大到小",sc)
sc.sort()
print("由小到大",sc)
```
