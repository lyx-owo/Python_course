假設班上10位同學的數學成績是90,80,88,84,85,87,88,89,100

請幫第一位同學（索引值0）的同學加10分

請幫最後一位同學（索引值-1）的同學減10分


```python
classmate = [90,80,88,84,85,87,88,89,100]
classmate[0]=classmate[0]+10
print("第一位同學新的成績", classmate)
```
