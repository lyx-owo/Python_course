串列綜合練習

1.請先產生score串列，內含80,90,85,75,70,65,88,98,100

2.請將75分的成績改成78分

3.請印出score的最高分，最低分及總分

4.在score串列尾端分別加上50,99，印出score串列

5.由小到大印出score串列

6.由大到小印出score串列

```python
score=[80,90,85,75,70,65,88,98,100]
score[3]=78
print('目前最高分=',max(score))
print('目前最低分=',min(score))
print('目前總分=',sum(score))
score.append(50)
score.append(99)
print('目前串列內容=',score)
score.sort()
print('由小排到大=',score)
score.sort(reverse=True)
print('由大排到小=',score)
```
