```python
year = input("請輸入出生年份: ")
yr=int(year)
rem = (yr-1900)%12
if (rem==0):
    print("鼠")
elif (rem==1):
    print("牛")
elif (rem==2):
    print("虎")
elif (rem==3):
    print("兔")
elif (rem==4):
    print("龍")
elif (rem==5):
    print("蛇")
elif (rem==6):
    print("馬")
elif (rem==7):
    print("羊")
elif (rem==8):
    print("猴")
elif (rem==9):
    print("雞")
elif (rem==10):
    print("狗")
else:
    print("豬")
```
