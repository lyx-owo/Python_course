```python
score = [80, 90, 85, 75]
print('目前串列內容 = ' ,score)
score[-1] = 78
print('修改後串列內容 = ' ,score)
print('最高分 = ' , max(score))
print('最低分 = ' , min(score))
print('總分 = ' , sum(score))
score.append (50)
score.append (99)
print('修改後串列內容 = ' ,score)
score.sort()
print('小到大排序 = ' ,score)
score.sort(reverse=True)
print('大到小排序 = ' ,score)
```
